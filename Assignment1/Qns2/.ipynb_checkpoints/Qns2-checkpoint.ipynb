2.	Compare the accuracy of Logistic Regression before and after applying PCA. 

```python
# IMPORT THE REQUIRED LIBRARIES

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# IMPORT THE DATASET: "diabetes.csv"

data = pd.read_csv("diabetes.csv")
```

```python
# PRINT THE FIRST 5 ROWS AND LAST 5 ROWS

print("FIRST 5 ROWS")
print(data.head(),"\n")

print("LAST 5 ROWS")
print(data.tail(),"\n")
```

```python
# ASSIGNING THE INPUT VALUE(x) AND OUTPUT VALUE (y)

x = data.iloc[:,:-1]
print(x,"\n")

y = data.iloc[:,-1]
print("Output\n")
print(y)
```

```python
# SPLIT THE DATASET INTO THE TRAINING DATA AND TESTING DATA

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=1/3,random_state =0)
```

```python
#MODEL TRAINING USING LOGISTIC REGRESSION

from sklearn.linear_model import LogisticRegression
reg=LogisticRegression(max_iter = 500)
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
print("Output\n",y_pred)
```

```python
# REQUIREMENTS FOR PCA

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

#STANDARDIZE FEATURES

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

#DIMENSIONALITY REDUCTION USING PCA
pca = PCA(n_components = 2)
x_train_pca = pca.fit_transform(x_train)
x_test_pca= pca.transform(x_test)
```

```python
#MODEL TRAINING USING LOGISTIC REGRESSION USING PCA

from sklearn.linear_model import LogisticRegression
reg_pca=LogisticRegression(max_iter = 500)
reg_pca.fit(x_train_pca,y_train)
```

```python
y_pred_pca = reg_pca.predict(x_test_pca)
print("Output\n",y_pred_pca)
```

```python
# ACCURACY OF LOGISTIC REGRESSION USING PCA AND WITHOUT PCA 

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy for LOGISTIC REGRESSION using WITHOUT PCA")
print(accuracy)

accuracy_pca = accuracy_score(y_test,y_pred_pca)
print("Accuracy for LOGISTIC REGRESSION using WITH PCA")
print(accuracy_pca)
```

```python
#CONFUSION MATRIX OF LOGISTRIC REGRESSION FOR PCA AND WITHOUT PCA

from sklearn.metrics import confusion_matrix

a = confusion_matrix(y_test,y_pred)
print("Confusion Matrix without PCA\n",a)

b = confusion_matrix(y_test,y_pred_pca)
print("Confustion Matrix with PCA\n",b)

```

```python
#CONFUSION MATRIX DISPLAY OF LOGISTIC REGRESSION FOR PCA AND WITHOUT PCA

from sklearn.metrics import ConfusionMatrixDisplay

display = ConfusionMatrixDisplay(confusion_matrix = a,display_labels = reg.classes_)
display.plot()
plt.title("CONFUSION MATRIX for WITHOUT PCA")
plt.show()

display_pca = ConfusionMatrixDisplay(confusion_matrix = b,display_labels = reg_pca.classes_)
display.plot()
plt.title("CONFUSION MATRIX for WITH PCA")
plt.show()
```
