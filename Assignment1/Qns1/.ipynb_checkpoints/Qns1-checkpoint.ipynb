1.	Implement PCA on the Iris dataset and visualize the data before and after dimensionality reduction (2D scatter plot).

```python
# IMPORT THE REQUIRED PYTHON LIBRARIES
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# LOAD THE IRIS DATASET
data = load_iris()
x, y = data.data, data.target
```

```python
# Split the dataset into training and testing data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# STANDARDIZE FEATURES
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# WITHOUT PCA - Use only first 2 features for visualization

x_train_2d = x_train[:, :2]  
x_test_2d = x_test[:, :2]
```

```python
# WITH PCA - Apply PCA to reduce to 2 dimensions
pca = PCA(n_components=2)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)

print(f"Explained variance ratio: {pca.explained_variance_ratio_}")
print(f"Total variance explained: {sum(pca.explained_variance_ratio_):.2%}")
```

```python
# Train classifiers on 2D data (first 2 features only)
rf_clf_no_pca = RandomForestClassifier(random_state=42)
rf_clf_no_pca.fit(x_train_2d, y_train)

ada_clf_no_pca = AdaBoostClassifier(random_state=42)
ada_clf_no_pca.fit(x_train_2d, y_train)

print("WITHOUT PCA - Training completed!")
```

```python

# Train classifiers on PCA-reduced data
rf_clf_pca = RandomForestClassifier(random_state=42)
rf_clf_pca.fit(x_train_pca, y_train)

ada_clf_pca = AdaBoostClassifier(random_state=42)
ada_clf_pca.fit(x_train_pca, y_train)

print("WITH PCA - Training completed!")
```

```python
# Function to plot decision boundaries
def plot_decision_boundaries(clf, X, y, ax, title):
    h = 0.02  # Step size in the mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    ax.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
    scatter = ax.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', marker='o', 
                        cmap='viridis', s=50, alpha=0.8)
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.set_title(title)
    ax.grid(True, alpha=0.3)
    return scatter
```

```python
# Create plots for WITHOUT PCA
fig, axs = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle('Decision Boundaries WITHOUT PCA (First 2 Features Only)', fontsize=16, fontweight='bold')

# Plot Random Forest
scatter1 = plot_decision_boundaries(rf_clf_no_pca, x_test_2d, y_test, axs[0], 
                                    'Random Forest (No PCA)')

# Plot AdaBoost
scatter2 = plot_decision_boundaries(ada_clf_no_pca, x_test_2d, y_test, axs[1], 
                                    'AdaBoost (No PCA)')

# Add colorbar
plt.colorbar(scatter2, ax=axs, label='Class')
plt.tight_layout()
plt.show()

# Calculate accuracy WITHOUT PCA
rf_acc_no_pca = rf_clf_no_pca.score(x_test_2d, y_test)
ada_acc_no_pca = ada_clf_no_pca.score(x_test_2d, y_test)
print(f"\nAccuracy WITHOUT PCA:")
print(f"Random Forest: {rf_acc_no_pca:.2%}")
print(f"AdaBoost: {ada_acc_no_pca:.2%}")
```

```python
# Create plots for WITH PCA
fig, axs = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle('Decision Boundaries WITH PCA (2 Principal Components)', fontsize=16, fontweight='bold')

# Plot Random Forest
scatter1 = plot_decision_boundaries(rf_clf_pca, x_test_pca, y_test, axs[0], 
                                    'Random Forest (With PCA)')

# Plot AdaBoost
scatter2 = plot_decision_boundaries(ada_clf_pca, x_test_pca, y_test, axs[1], 
                                    'AdaBoost (With PCA)')

# Add colorbar
plt.colorbar(scatter2, ax=axs, label='Class')
plt.tight_layout()
plt.show()

# Calculate accuracy WITH PCA
rf_acc_pca = rf_clf_pca.score(x_test_pca, y_test)
ada_acc_pca = ada_clf_pca.score(x_test_pca, y_test)
print(f"\nAccuracy WITH PCA:")
print(f"Random Forest: {rf_acc_pca:.2%}")
print(f"AdaBoost: {ada_acc_pca:.2%}")
```
